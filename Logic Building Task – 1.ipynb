```python
# 1. USER LOGIN CHECK

def user_check(username, password):
    stored_username = "admin"
    stored_password = "1234"

    if username == stored_username and password == stored_password:
        print("Login successful!")
    else:
        print("Invalid Credentials")


# USER INPUT
username = input("Enter your username: ")
password = input("Enter your password: ")

user_check(username, password)

```

```python
# Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

passed_students = 0
failed_students = 0

for score in marks:
    if score >= 50:
        passed_students += 1
    else:
        failed_students += 1

print("===== Result Summary =====")
print("Total Students:", len(marks))
print("Number of Students Passed:", passed_students)
print("Number of Students Failed:", failed_students)
```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]

print("Cleaned List:", cleaned_names)

```

```python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for text in messages:
    character_count = len(text)
    
    print("Message:", text)
    print("Character Count:", character_count)
    
    if character_count > 10:
        print("Alert: This message exceeds 10 characters.")
    else:
        print("Info: This message is within the limit.")
    
    print()
```
